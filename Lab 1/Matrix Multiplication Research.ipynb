```python
import csv
import numpy
import matplotlib.pyplot as plt 

def read_data(file_path):
    matrix_size = list()
    mult_simple = list()
    mult_block = list()
    mult_swap = list()
    
    with open(file_path, newline='') as file:
        reader = csv.reader(file, delimiter=',', quotechar='|')
        for row in reader:
            matrix_size.append(int(row[0]))
            mult_simple.append(float(row[1]))
            mult_block.append(float(row[2]))
            mult_swap.append(float(row[3]))
            
    return (matrix_size, mult_simple, mult_block, mult_swap)

def display_plot_from_file(filename):
    (matrix_size, mult_simple, mult_block, mult_swap) = read_data(filename)

    plt.plot(matrix_size, mult_simple, 'r', matrix_size, mult_block, 'g', matrix_size, mult_swap, 'b')
    plt.xlabel('Matrix Size')
    plt.ylabel('Time')
    plt.show()
```

```python
display_plot_from_file("1000_1100_float.csv")
```

```python
display_plot_from_file("1000_1100_double.csv")
```

```python
display_plot_from_file("1000_2000_float.csv")
```
